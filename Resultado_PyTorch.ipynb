```python
# with open(f'    drive/MyDrive/Colab Notebooks/dados_artigo/result_{Target}.txt', 'rb') as f:
with open(f'result_{Target}.txt', 'rb') as f:
  results = pickle.load(f)

rmses6 = {}
rmses1 = {}

for i in results:
  if (results[i]['window_size'] == 1):
    rmses1[int(i.split('.')[0])] = list(results[i]['error'])
  else:
    rmses6[int(i.split('.')[0])] = list(results[i]['error'])

rmses1 = pd.DataFrame(rmses1)
rmses6 = pd.DataFrame(rmses6)
```

```python
df_1M = pd.DataFrame()

df_1M['RMSE'] = pd.DataFrame(rmses1).apply(lambda column : mean_squared_error(column, np.zeros(len(column)), squared=False))
df_1M['MAE'] = pd.DataFrame(rmses1).apply(lambda column : mean_absolute_error(column, np.zeros(len(column))))
df_1M['Modelo'] = rmses1.apply(lambda row : 'PyTorch LSTM'if row.name >= 13 else 'LSTM', axis=1 )
df_1M['Meses'] = 1
df_1M = df_1M[['Modelo','Meses', 'RMSE', 'MAE']]
df_1M
# df_1M.sort_values(['RMSE', 'MAE'])
# df_1M.sort_values(['MAE', 'RMSE'])
```

```python
df_6M = pd.DataFrame()

df_6M['RMSE'] =rmses6.apply(lambda column : mean_squared_error(column, np.zeros(len(column)), squared=False))
df_6M['MAE'] =rmses6.apply(lambda column : mean_absolute_error(column, np.zeros(len(column))))
df_6M['Modelo'] = rmses1.apply(lambda row : 'PyTorch LSTM' if row.name >= 13 else 'LSTM', axis=1 )
df_6M['Meses'] = 6
df_6M = df_6M[['Modelo','Meses', 'RMSE', 'MAE']]
df_6M
# df_6M.sort_values(['RMSE', 'MAE'])
# df_6M.sort_values(['MAE', 'RMSE'])
```

```python
best_models = df_1M.query("Modelo == 'PyTorch LSTM'").sort_values(['RMSE', 'MAE']).head(1)
best_models = pd.concat([best_models, df_6M.query("Modelo == 'PyTorch LSTM'").sort_values(['RMSE', 'MAE']).head(1)])
best_models
```
